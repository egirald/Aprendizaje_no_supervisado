```python
from category_encoders import BinaryEncoder

encoder = BinaryEncoder(cols=["Codigo de Zona"])

# Ajustamos el codificador con native-country y la transformamos
encoder.fit(datos["Codigo de Zona"])
df_binario = encoder.transform(datos["Codigo de Zona"])
datos = pd.concat([datos, df_binario], axis=1)
```

```python
datos.drop(columns=['Codigo de Zona'], inplace=True)
datos
```

```python
encoder2 = BinaryEncoder(cols=["Modelo del hardware"])

# Ajustamos el codificador con native-country y la transformamos
encoder2.fit(datos["Modelo del hardware"])
df_binario2 = encoder2.transform(datos["Modelo del hardware"])
datos = pd.concat([datos, df_binario2], axis=1)
datos.drop(columns=['Modelo del hardware'], inplace=True)
datos
```

```python
datos.set_index('Nombre', inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
datos_escalados = scaler.fit_transform(datos)
datos_escalados
```

```python
datos_escalados_df = pd.DataFrame(data = datos_escalados, index=datos.index, columns=datos.columns)
datos_escalados_df
```

## Componentes principales

```python
## valores singulares y vectores

U, S, Vt = np.linalg.svd(datos_escalados_df, full_matrices=False)

print(U.shape)
print(S.shape)
print(Vt.shape)

var_explicada = (S**2/np.sum(S**2))

for k in range(10,30):
    print(k,": ", np.round([sum(var_explicada[:k])],2))
```

## K-medias

```python
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score

varianza_intra_cluster_kmeans = []
silhouettes_kmeans = []

for k in range (1, 10):
    kmeans_t = KMeans(n_clusters = k, random_state = 1234).fit(datos_escalados_df)
    varianza_intra_cluster_kmeans.append(kmeans_t.inertia_)
    try:
        silhouette_kmeans_t = silhouette_score(datos_escalados_df, kmeans_t.labels_)
    except:
        silhouette_kmeans_t = 0 # El índice de Silhouette sólo se puede calcular cuando se tiene K>1
    silhouettes_kmeans.append(silhouette_kmeans_t)
```

```python
plt.plot(range(1, 10), varianza_intra_cluster_kmeans, marker='o')
plt.xlabel('Número de clústeres (K) k-means')
plt.ylabel('Varianza intra clúster')
plt.show()

plt.plot(range(1, 10), silhouettes_kmeans, marker='o')
plt.xlabel('Número de clústeres (K) k-means')
plt.ylabel('Índice de Silhouette')
plt.show()
```
